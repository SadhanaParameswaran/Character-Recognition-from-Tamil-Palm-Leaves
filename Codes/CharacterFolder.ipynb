```python
import cv2
import os

# Define input and output folders
input_folder = "D://AMRITA COLLEGE//SEM 6//Deep Learning//Project2//6"
output_folder = "D:\\AMRITA COLLEGE\\SEM 6\\Deep Learning\\Project2\\6_output"

# Create output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Loop over each image in the input folder
for filename in os.listdir(input_folder):
    # Load the image
    img_path = os.path.join(input_folder, filename)
    img = cv2.imread(img_path)

    # Convert to grayscale
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Apply thresholding
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU)[1]

    # Find contours
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Perform connected component analysis
    num_labels, labels, stats, centroids = cv2.connectedComponentsWithStats(thresh)

    # Loop over each connected component and extract the character
    for i in range(1, num_labels):
        # Get the bounding box of the connected component
        (x, y, w, h, area) = stats[i]

        # Check if the character is large enough
        if w > 10 and h > 10:
            # Extract the character by cropping the image
            char = img[y:y+h, x:x+w]

            # Save the character image
            output_path = os.path.join(output_folder, 'char_{}_{}'.format(filename[:-10], i) + '.jpg')
            cv2.imwrite(output_path, char)

        else:
            # Delete the small character image
            os.remove(os.path.join(output_folder, 'char_{}_{}'.format(filename[:-10], i) + '.jpg'))

```
