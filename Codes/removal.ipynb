```python
from PIL import Image
import os
import time

# Set the path of the folder containing the images
folder_path = 'D://AMRITA COLLEGE//SEM 6//Deep Learning//Project2//7_output'

# Loop through each file in the folder
for filename in os.listdir(folder_path):
    filepath = os.path.join(folder_path, filename)
    
    # Wait for any file locks to be released
    while True:
        try:
            with Image.open(filepath) as img:
                break
        except:
            time.sleep(0.1)
    
    # Get the dimensions of the image
    width, height = img.size
    
    # Check if the dimensions are less than 15x15
    if width < 30 or height < 30:
        
        # If the dimensions are less than 15x15, delete the file
        os.remove(filepath)
        
        # Print a message to confirm that the file was deleted
        print(f"{filename} deleted")

```
